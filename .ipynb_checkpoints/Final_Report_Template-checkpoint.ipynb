## 1. Analyzing global CO2 trends  
<!--(Replace "Section 1 Title" with something meaningful, e.g., "Analyzing Seasonal Temperature Trends")* 

In this cell, introduction the data set(s) you are working with, specify which aspect of your research question this section addresses, and describe the figure you will produce. -->

The purpose of this section is to get a trend of overall CO2 in the atmosphere from 1974. It shows that CO2 has been consistently rising for the last 50 years.

```python
# load data file(s)
co2_df=pd.read_csv('co2_mm_mlo.csv',usecols=['year','month','average'])

# as part of commenting your code throughout this section, discuss any problems you encountered and how you solved them
```

```python
# perform data cleaning steps

co2_df['year']=pd.to_numeric(co2_df['year'])
co2_1980=co2_df[np.logical_and(co2_df['year']>1979, co2_df['year']<2024)]

Year = pd.to_datetime(co2_1980.year,format='%Y').dt.year
Month = pd.to_datetime(co2_1980.month,format='%m').dt.month
co2_df_datetime = pd.to_datetime(co2_1980[['year', 'month']].assign(day=1))
```

```python
# perform data analysis

display(co2_1980)
x=co2_df_datetime
y=co2_1980['average']
```

```python
# make and display the first figure

fig, axs = plt.subplots(figsize=(16,5))

axs.plot(x,y, marker='.',c='red')
axs.set_xlabel('Time', fontsize=15)
axs.set_ylabel("CO2 (ppm)", fontsize=15)
axs.set_title('Time vs. CO2', fontsize=20)
axs.grid()
```

<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 1 Caption:** Change in CO2 from 1980-2024. This figure shows an increase of almost 100 ppm over this time period.

## 2. Analyzing changes in Ice Extent over time  

<!-- In the introduction below, specify the data set(s) you are working with in this section, which aspect of your research question this section addresses, and the figure you will produce -->

This section uses dataset 2 to analyze the change in yearly ice extent since 1979. The figure will show ice extent in millions of km^2 over time.

```python
# if a new dataset, load data and perform data cleaning steps
# otherwise, continue to data analysis and figure creation
ice_extent_df=pd.read_csv('N_01_extent_v3.0.csv')
ice_extent_df[' extent'] = ice_extent_df[' extent'].replace(-9999, np.nan)
ice_1980=ice_extent_df[ice_extent_df['year']>1979]
```

```python
# perform data analysis
x2=ice_1980['year']
y2=ice_1980[' extent']
```

```python
# make and display figure
fig, axs = plt.subplots(figsize=(16,5))
#time v ice extent

axs.plot(x2,y2, marker='.',c='blue')
axs.set_xlabel('Time', fontsize=15)
axs.set_ylabel("Ice Extent (millions of km^2)", fontsize=15)
axs.set_title('Time vs. Ice Extent', fontsize=20)
axs.grid()
```

<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 2 Caption:** This figure shows the overall change in ice extent each year since 1980. You can see that ice coverage has been decreasing over this time period.

## 3. Linear Regressions

<!-- In the introduction below, specify the data set(s) you are working with in this section, which aspect of your research question this section addresses, and the figure you will produce -->

This section uses both datasets to find the linear regressios of both the sea ice and CO2 trends. This helps compare the two plots to eachother and determine how they relate to eachother.

```python
# if a new dataset, load data and perform data cleaning steps
# otherwise, continue to data analysis and figure creation
```

```python
# perform data analysis
#linear regression
# Convert year and month to datetime
ice_1980['year']=pd.to_numeric(ice_1980['year'])

co2_df_datetime = pd.to_datetime(co2_1980[['year', 'month']].assign(day=1))

#linear regression for co2
slope, intercept, rvalue, pvalue, stderr = stats.linregress(co2_df_datetime.astype(int), co2_1980['average'])

#linear regression fit
co2_1980.loc[:, 'regression']=co2_df_datetime.astype(int)*slope+intercept

ice_1980_notna = ice_1980.dropna()
 
#linear regression for ice
slope2, intercept2, rvalue2, pvalue2, stderr2 = stats.linregress(ice_1980_notna['year'], ice_1980_notna[' extent'])

ice_1980.loc[:, 'regression']=ice_1980['year']*slope2+intercept2
print('CO2 Trendline')
print('Slope=', slope, 'R^2=',rvalue**2,'P value=',pvalue)
print('Ice Extent Trendline')
print('Slope=', slope2, 'R^2=',rvalue2**2,'P value=',pvalue2)
```

```python
# make and display figure
# Create the figure and axis
fig, ax1 = plt.subplots(figsize=(16,5))

# Plot CO2 levels on the left y-axis
x_co2 = co2_df_datetime
y_co2 = co2_1980['average']
ax1.plot(x_co2, y_co2, marker='.', c='red', label='CO2 (ppm)')
ax1.plot(x_co2,co2_1980['regression'],linestyle='--',c='maroon', label='CO2 Linear Regression')
ax1.set_xlabel('Time', fontsize=15)
ax1.set_ylabel('CO2 (ppm)', fontsize=15, color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for ice extent
ax2 = ax1.twinx()
x_ice = pd.to_datetime(ice_1980['year'], format='%Y')
y_ice = ice_1980[' extent']
ax2.plot(x_ice, y_ice, marker='.', c='blue', label='Ice Extent (millions of km^2)')
ax2.plot(x_ice,ice_1980['regression'],linestyle='--',c='darkblue', label='Ice Linear Regression')
ax2.set_ylabel('Ice Extent (millions of km^2)', fontsize=15, color='blue')
ax2.tick_params(axis='y', labelcolor='darkblue')

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper center', bbox_to_anchor=(.5, 1))

# Set title and grid
plt.title('Time vs. CO2 and Ice Extent', fontsize=20)
ax1.grid()

# Show the plot
plt.show()
```

<!-- Describe how this figure helps address your research question. What patterns or insights can you observe? -->
**Figure 3 Caption:** This figure shows the liner regressions of the two graphs above and demonstrates an inversely related trend between ice extent and CO2.
